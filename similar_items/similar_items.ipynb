```python
import json
import csv
import itertools
import pandas as pd
import numpy as np
```

```python
female_lookup = open("parent_item_data_female_26-02-2018.json")

look_up_f = json.load(female_lookup)

male_lookup = open("parent_item_data_male_26-02-2018.json")

look_up_m = json.load(male_lookup)


# look_up = look_up_f.copy()   # start with x's keys and values
# look_up.update(look_up_m)
```

```python
male_cm_ic = open("men_ic_correlation.json")
look_up_male_cm_ic = json.load(male_cm_ic)

female_cm_ic = open("women_ic_correlation.json")
look_up_female_cm_ic = json.load(female_cm_ic)

male_cm_brand = open("men_brand_correlation.json")
look_up_male_cm_brand = json.load(male_cm_brand)

female_cm_brand = open("women_brand_correlation.json")
look_up_female_cm_brand = json.load(female_cm_brand)

male_cm_color = open("men_color_correlation.json")
look_up_male_cm_color = json.load(male_cm_color)

female_cm_color = open("women_color_correlation.json")
look_up_female_cm_color = json.load(female_cm_color)
```

```python
look_up_m
```

```python
# Men item similarity

# LS = BS + PS1*CFinventory_cat + PS2*CFcolor_tone + PS3*CFbrand_id
PSIC = 60 # inventory category preference score
PSB = 20 # brand preference score
PSC = 20 # color preference score
base_parent_item = "FRES10845"
base_brand_id = str(445)
base_color_id = str(28)
base_inventory_category_id = str(36)
similar_items_list = {base_parent_item: {}}
for key,val in look_up_m.items():
    if key != base_parent_item and key not in similar_items_list[base_parent_item].keys():
        similar_items_list[base_parent_item][key] = {
            'inv_cat': 0,
            'brand': 0,
            'color': 0
        }
        
        try:
            similar_items_list[base_parent_item][key]['inv_cat'] = PSIC * look_up_male_cm_ic[base_inventory_category_id][str(val['inventory_category'])] 
            print()
        except KeyError:
            pass
        
        try:
            similar_items_list[base_parent_item][key]['brand'] = PSB * look_up_male_cm_brand[base_brand_id][str(val['brand_id'])] 
        except KeyError:
            pass
        
        try:
            similar_items_list[base_parent_item][key]['color'] = PSC * look_up_male_cm_color[base_color_id][str(val['color'])]
        except KeyError:
            pass
        
        similar_items_list[base_parent_item][key]['link_score'] = similar_items_list[base_parent_item][key]['inv_cat'] + similar_items_list[base_parent_item][key]['brand'] + similar_items_list[base_parent_item][key]['color']
        
        
```

```python
similar_items_list
```

```python
with open('men_similar_items.json', 'w') as outfile:
    json.dump(similar_items_list, outfile)
```
