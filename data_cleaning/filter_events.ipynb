```python
import json
import csv
import pandas as pd
import glob
```

```python
path =r'/home/flyrobe/Documents/elasticsearch_to_json/add_to_cart_web/' # use your path
allFiles = glob.glob(path + "/*.json")
print(allFiles)
add_to_cart = []
for file_ in allFiles:
    #df = pd.read_json(file_)
    json_data = open(file_)
    jdata = json.load(json_data)
    initial_list = [each["_source"] for each in jdata if each["_source"]["event"] == "Add to Cart Web"]
    add_to_cart += initial_list
path =r'/home/flyrobe/Documents/elasticsearch_to_json/details_web/' # use your path
allFiles = glob.glob(path + "/*.json")
print(allFiles)
details_web = []
for file_ in allFiles:
    #df = pd.read_json(file_)
    json_data = open(file_)
    jdata = json.load(json_data)
    initial_list = [each["_source"] for each in jdata if each["_source"]["event"] == "Details Web"]
    details_web += initial_list
    

final_list = add_to_cart + details_web
#event_sample = pd.DataFrame(final_list)
```

```python
with open('/home/flyrobe/Documents/elasticsearch_to_json/bronze/data.json', 'w') as outfile:
    json.dump(final_list, outfile)
```
